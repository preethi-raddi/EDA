```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
Z = linkage(data, method='ward')
plt.figure(figsize=(10, 6))
dendrogram(Z, labels=data['eco1'].values, leaf_rotation=90, leaf_font_size=10)
```

```python
from sklearn.cluster import KMeans
```

```python
data_kmeans = data_eda[['political.knowledge', 'Europe']]
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data_kmeans)
```

```python
sns.scatterplot(data=data_kmeans, x='political.knowledge', y='Europe')
plt.title("21BML0109")
```
